```python
# Electricity generation by source

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns

country = [['SLV','PAN','CHL','DNK', 'DEU', 'FRA', 'GBR','USA', 'JPN'],['IND', 'CHN', 'BRA', 'CRI'], ['ETH', 'RWA', 'KEN'], \
          ['IDN', 'KHM', 'BRN', 'THA', 'SGP', 'LAO', 'PHL', 'VNM', 'MYS', 'MMR'], ['LKA', 'PAK', 'UZB', 'TJK']]
colors = ["gray", "silver", "black","salmon","steelblue","yellow", "skyblue","orange","green"]
labels = ['oil','gas','coal','nuclear','hydro','bioenergy', 'wind','solar','otherenewable']

df = pd.read_csv('/Users/kurozuhajime/Desktop/Pythonコード・データ/電力源/goal07.electricitysources.csv')

x = range(2000, 2021+1)

for catgry in country:
    for i, ctry in enumerate(catgry, 1):
        data = df[df['iso3c'] == ctry]
        #fig = plt.subplot()
        plt.title(ctry)
        plt.stackplot(x, [data['oil']/1000, data['gas']/1000, data['coal']/1000, data['nuclear']/1000, data['hydro']/1000, data['bioenergy']/1000, \
                    data['wind']/1000, data['solar']/1000,data['otherrenewable']/1000], colors=colors, labels=labels)
        plt.xlabel('year')
        plt.ylabel('TWh')
        plt.xticks(x[::3], rotation=0)
        #plt.legend(fontsize=8, bbox_to_anchor=(0.8, -0.12), ncol=3)
        plt.legend(labelcolor="k",facecolor='w', edgecolor='w',loc='upper left', fontsize=7)
        plt.show()

```
