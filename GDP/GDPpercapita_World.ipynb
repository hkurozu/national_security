```python
import pandas as pd
import matplotlib.pyplot as plt

gdp_per_capita = pd.read_csv('/Users/kurozuhajime/Desktop/Pythonコード・データ/GDP/GDPdata/NGDPDPC-1-表1.csv')

plt.title('Per capita GDP', fontsize=10)
plt.plot(gdp_per_capita['year'], gdp_per_capita['Advanced'], color='k', label='advanced')
plt.plot(gdp_per_capita['year'], gdp_per_capita['World'], color='k', linestyle='--', label='world')
plt.plot(gdp_per_capita['year'], gdp_per_capita['Developing'], color='k', linestyle=':',label='developing')

plt.xticks(gdp_per_capita['year'][::4], fontsize=10)
plt.yticks(plt.yticks()[0], ["{:,}".format(int(i)) for i in plt.yticks()[0]])
plt.ylim([0,60000])
plt.xlabel('year', fontsize=10)
plt.ylabel('US$', fontsize=10)
plt.legend(fontsize=8)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/Users/kurozuhajime/Desktop/Pythonコード・データ/GDP/GDPdata/NGDPDPC-表1.csv')

list = range(1980, 2024)
data = df.pivot_table([ str(l) for l in list], index='category')
#print(data)

advc = data.iloc[0, :]
#print(advc)
devp = data.iloc[1, :]
wrld = data.iloc[2, :]

plt.title('Per capita GDP', fontsize=10)
plt.plot(list, advc, color='k', label='advanced')
plt.plot(list, wrld, color='k', linestyle='--', label='world')
plt.plot(list, devp, color='k', linestyle=':',label='developing')

plt.xticks(gdp_per_capita['year'][::4], fontsize=10)
plt.yticks(plt.yticks()[0], ["{:,}".format(int(i)) for i in plt.yticks()[0]])
plt.ylim([0,60000])
plt.xlabel('year', fontsize=10)
plt.ylabel('US$', fontsize=10)
plt.legend(fontsize=8)

```
