```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/Users/kurozuhajime/Desktop/Pythonコード・データ/防災予算/防災データ/シート1-siryo33_2.csv')
print(df.columns)


fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

plt.title('Budget for disaster prevention, Japan', fontsize=10)

ax1.bar(df['year'], df['科学技術の研究'], width=0.8, alpha=0.6, color='darkgoldenrod', label='research in Science and Technology')
ax1.bar(df['year'], df['災害予防'], width=0.8, alpha=0.6, color='Navy',label='disaster prevention', bottom=df['科学技術の研究'])
ax1.bar(df['year'], df['国土保全'], width=0.8, alpha=0.6, color='steelblue', label='land conservation', bottom=df['科学技術の研究']+df['災害予防'])
ax1.bar(df['year'], df['災害復旧等'], width=0.8, alpha=0.6, color='green', label='disaster recovery, etc.', bottom=df['科学技術の研究']+df['災害予防']+df['国土保全'])

ax2.plot(df['year'], df['対比'], color='darkred', label='a share (%)')

plt.xticks(df['year'][::6], rotation=90)

ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
#ax2.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

ax1.set_ylim(0, 8800000)
ax2.set_ylim(0, 12.8)

ax1.set_xlabel('year')
ax1.set_ylabel('Budgets for disaster prevention (million yen)')
ax2.set_ylabel('a share of the general account budget (%)')

handler1, label1 = ax1.get_legend_handles_labels()
handler2, label2 = ax2.get_legend_handles_labels()
ax1.legend(handler1 + handler2, label1 + label2, labelcolor="k",facecolor='w', edgecolor='w', framealpha=0.8, ncol=3, loc='upper center', borderaxespad=0., fontsize=7)

plt.show()
```
